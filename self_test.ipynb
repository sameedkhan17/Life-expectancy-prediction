```python
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
life_exp['Status']= label_encoder.fit_transform(life_exp['Status'])
life_exp['Country']= label_encoder.fit_transform(life_exp['Country'])
```

```python
plt.figure(figsize=(20, 18))    
sns.heatmap(life_exp.corr(),annot=True)
plt.show()
```

```python
pd.set_option('display.max_columns', None)
life_exp.head()
```

```python
life_exp.columns = life_exp.columns.str.lstrip()
life_exp.columns = life_exp.columns.str.rstrip()
```

```python
life_exp.columns
```

```python
sns.pairplot(life_exp,x_vars=['Year','Status','Adult Mortality','infant deaths','Alcohol','percentage expenditure','Hepatitis B','Measles','BMI','under-five deaths','Polio','Total expenditure','Diphtheria','HIV/AIDS','GDP','Population','Income composition of resources','Schooling'],y_vars='Life expectancy',size=4,kind='scatter')
plt.show()
```

```python
X = life_exp[['Adult Mortality','infant deaths','Alcohol','percentage expenditure','BMI','under-five deaths','Polio','Diphtheria','HIV/AIDS','GDP','thinness  1-19 years','thinness 5-9 years','Income composition of resources','Schooling']]
y = life_exp['Life expectancy']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, train_size = 0.7, test_size = 0.3, random_state = 100 )
```

```python
print(X_test)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)

# 10. Evaluate the model
from sklearn.metrics import mean_squared_error, r2_score

print("Intercept:", lr.intercept_)
print("Coefficient:", lr.coef_)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R-squared:", r2_score(y_test, y_pred))
```

```python
# 1. Open a file in write-binary mode
with open('linear_model.pkl', 'wb') as f:
    # 2. Dump your fitted model object to that file
    pickle.dump(lr, f)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
with open('linear_model.pkl', 'rb') as f:
    lr = pickle.load(f)

# 3) Predict on the test set
y_pred = lr.predict(X_test)
```
